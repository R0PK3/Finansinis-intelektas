**II-III — Savarankiška duomenų paieška ir užkrovimas į Python**




Importai

```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
import math
```

Funkcijos

```python
def readFile(data_dir, size):
  data = pd.read_csv(data_dir, parse_dates=['Local time'], dayfirst=True)
  data = data.set_index('Local time')
  data_printer(data)
  return data.head(size)

def data_printer(data):
  print(data.dtypes)
  print(data)


def drawGraph(data, column_names, x_axis, y_axis, title, interval, format,step):
  plt.figure(figsize = (15,7))

  for column_name in column_names:
    plt.plot(data.index, data[column_name], label = column_name)

  plt.xlabel(x_axis)
  plt.ylabel(y_axis)
  plt.title(title)
  plt.legend()
  plt.grid(color = 'green', linestyle = '--')

  plt.xlim(data.index[0], data.index[-1])
  plt.gca().xaxis.set_major_locator(interval)
  plt.gca().xaxis.set_major_formatter(format)

  for column_name in column_names:
    max_value = data[column_name].max()
    min_value = data[column_name].min()

  plt.yticks(np.arange(math.floor(min_value), math.ceil(max_value), step = step))

  plt.gcf().autofmt_xdate()
  plt.show()
```

Tikiniai duomenys

```python
data = readFile('ticks_bmw.csv', 300)

data['TickNo'] = np.array(range(1,301))
data = data.set_index('TickNo')

drawGraph(data = data,
          column_names = ['Ask', 'Bid'],
          x_axis = 'Ticku skaicius',
          y_axis = 'Kaina €',
          title = 'BMW.DE/EUR Bayerische Motoren Werke AG, Tickiniai duomenys',
          interval = ticker.MultipleLocator(base = 20),
          format = ticker.FormatStrFormatter('%d'),
          step = 0.1)
```

Minutiniai duomenys

```python
data = readFile('minute_bmw.csv', 31)

drawGraph(data = data,
          column_names = ['Open', 'High', 'Low', 'Close'],
          x_axis = 'Laikas HH:MM nuo 20:00 iki 20:30',
          y_axis = 'Kaina €',
          title = 'BMW.DE/EUR Bayerische Motoren Werke AG, Minutiniai duomenys',
          interval = dates.MinuteLocator(interval = 1),
          format = dates.DateFormatter('%H:%M'),
          step = 0.05)
```

Dieniniai duomenys

```python
data = readFile('day_bmw.csv', 30)


drawGraph(data = data,
          column_names = ['Open', 'High', 'Low', 'Close'],
          x_axis = 'Data (MM-DD)',
          y_axis = 'Kaina €',
          title = 'BMW.DE/EUR Bayerische Motoren Werke AG, Dieniniai duomenys',
          interval = dates.DayLocator(interval = 1),
          format = dates.DateFormatter('%m-%d'),
          step = 0.5)
```

**Iš tikinių duomenų suformuoti valandinius barus**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

data = pd.read_csv('ticks_bmw.csv')
data['Local time'] = pd.to_datetime(data['Local time'])
data.set_index('Local time', inplace=True)

hourly_data = data.resample('H').agg({'Ask': ['first', 'max', 'min', 'last'],
                                            'Bid': ['first', 'max', 'min', 'last']})
print(hourly_data)

hourly_data.columns = ['Ask_Open', 'Ask_High', 'Ask_Low', 'Ask_Close',
                       'Bid_Open', 'Bid_High', 'Bid_Low', 'Bid_Close']

plot_data = hourly_data.head(10)
fig, ax = plt.subplots(figsize=(12, 6))

for i, row in plot_data.iterrows():
    ax.plot([i, i], [row['Ask_Low'], row['Ask_High']], color='black', linewidth=2)
    ax.plot([i, i - pd.Timedelta(hours=0.2)], [row['Ask_Open'], row['Ask_Open']], color='green', linewidth=2)
    ax.plot([i, i + pd.Timedelta(hours=0.2)], [row['Ask_Close'], row['Ask_Close']], color='red', linewidth=2)

ax.xaxis_date()
ax.xaxis.set_major_locator(mdates.HourLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax.grid(True)
ax.set_title('Valandiniai duomenys BMW.DE/EUR Bayerische Motoren Werke AG')
ax.set_xlabel('Laikas (HH:MM)')
ax.set_ylabel('Kaina €')
plt.tight_layout()
plt.show()
```
